#### <center> **1.3.1. ¿Cual es la función objetivo?** </center>

La función objetivo es aquella que comprueba y suma el valor de coste de los actores en base a si están presentes en ese día o no. El objetivo será el siguiente.

 $$\text{Minimizar: } \sum_{i \in A} \sum_{j \in D} x_{ij} \cdot \text{Costo}_{\text{diario}} \cdot \text{Actores\_por\_dia}[j][i]$$


- A es el conjunto de todos los actores.

- D es el conjunto de días de grabación.

- xij es una variable binaria que toma el valor 1 si el actor i está presente en el día j, y 0 en caso contrario.

- Costo diario es el costo diario (Constante en nuestor problema)

#### <center> **1.3.2.  ¿Es un problema de maximización o minimización?** </center>

Como bien observamos en el apartado anterior, se trara de un problema de minimización. Se trata de que xij salga como un 0 el mayor número de veces posibles, reduciendo la cantidad de veces que sumamos el coste de un día de un empleado.


#### <center> **1.4.1. Diseña un algoritmo para resolver el problema por fuerza bruta** </center>

El siguiente CODIGO FUNCIONA: PERO ES MUY MUY LENTO

```python
from itertools import combinations

def calculate_cost(num_tomas_per_day):
    return num_tomas_per_day * COST_PER_DAY

def find_optimal_schedule_bruteforce(num_actors, num_tomas, max_tomas_per_day, actors_per_toma):
    min_cost = float('inf')
    optimal_schedule = None

    # Generar todas las combinaciones posibles de tomas por día
    for r in range(1, min(max_tomas_per_day, num_tomas) + 1):
        for combination in combinations(range(num_tomas), r):
            # Verificar si los actores requeridos para esta combinación están presentes
            actors_required = set()
            for toma in combination:
                actors_required.update(actors_per_toma[toma])

            if len(actors_required) == num_actors:
                # Verificar los cuatro días siguientes sin repeticiones
                total_cost = calculate_cost(len(combination))
                days = [combination]
                used_tomas = set(combination)
                for i in range(1, 5):
                    next_combination = combinations(range(num_tomas), r + i)
                    for next_comb in next_combination:
                        if not any(toma in used_tomas for toma in next_comb):
                            next_actors_required = set()
                            for next_toma in next_comb:
                                next_actors_required.update(actors_per_toma[next_toma])

                            if len(next_actors_required) == num_actors:
                                total_cost += calculate_cost(len(next_comb))
                                days.append(next_comb)
                                used_tomas.update(next_comb)
                                break

                if total_cost < min_cost:
                    min_cost = total_cost
                    optimal_schedule = days

    return optimal_schedule, min_cost

# Datos de entrada (ejemplo)
NUM_ACTORS = 10
NUM_TOMAS = 30
MAX_TOMAS_PER_DAY = 15
COST_PER_DAY = 100  # Supongamos que el costo por día de cada actor es de 100

# Representación de las tomas con los actores requeridos (ejemplo)
actors_per_toma = actors_per_toma

optimal_schedule, min_cost = find_optimal_schedule_bruteforce(NUM_ACTORS, NUM_TOMAS, MAX_TOMAS_PER_DAY, actors_per_toma)

print("Planificación óptima de sesiones por día:")
for day, combination in enumerate(optimal_schedule, 1):
    print(f"Día {day}: {combination}")
print("Gasto mínimo total: $", min_cost)
```

### <center> **1.5.1. Diseña un algoritmo que mejore la complejidad del algortimo por fuerza bruta. Argumenta porque crees que mejora el algoritmo por fuerza bruta** </center>

Nuestro enfoque y resolución, será un algorimto genético. Otras opciones posibles como hemos descrito anteriormente, podría ser un algoritmo de backtracking con unas restricciones bien asentadas.

```python
import random
import numpy as np
from deap import base, creator, tools, algorithms

def calculate_cost(tomas_per_day):
    return len(tomas_per_day) * COST_PER_DAY

def evaluate(individual):
    cost = sum([calculate_cost(tomas_per_day) for tomas_per_day in individual])
    return cost,

def find_optimal_schedule(num_actors, num_tomas, max_tomas_per_day, num_days=4, population_size=100, generations=50):
    actors = list(range(1, num_actors + 1))
    tomas = list(range(num_tomas))

    # Representación del individuo: cada día es una lista de tomas
    creator.create("FitnessMin", base.Fitness, weights=(-1.0,))
    creator.create("Individual", list, fitness=creator.FitnessMin)

    # Crear caja de herramientas
    toolbox = base.Toolbox()
    toolbox.register("tomaset", random.sample, tomas, max_tomas_per_day)
    toolbox.register("individual", tools.initRepeat, creator.Individual, toolbox.tomaset, n=num_days)
    toolbox.register("population", tools.initRepeat, list, toolbox.individual)

    toolbox.register("evaluate", evaluate)
    toolbox.register("mate", tools.cxTwoPoint)
    toolbox.register("mutate", tools.mutShuffleIndexes, indpb=0.1)
    toolbox.register("select", tools.selTournament, tournsize=3)

    population = toolbox.population(n=population_size)
    hof = tools.HallOfFame(1)

    # Ejecutar el algoritmo genético
    algorithms.eaSimple(population, toolbox, cxpb=0.5, mutpb=0.2, ngen=generations, halloffame=hof)

    return hof[0], hof[0].fitness.values[0]

# Datos de entrada
NUM_ACTORS = 10
NUM_TOMAS = 30
MAX_TOMAS_PER_DAY = 6
COST_PER_DAY = 100  # Supongamos que el costo por día de cada actor es de 100

# Representación de las tomas con los actores requeridos
actors_per_toma = actors_per_toma

# Ejecutar la función para encontrar la planificación óptima de sesiones por día
optimal_schedule, min_cost = find_optimal_schedule(NUM_ACTORS, NUM_TOMAS, MAX_TOMAS_PER_DAY)

print("Planificación óptima de sesiones por día:")
print(optimal_schedule)
print("Gasto mínimo: $", min_cost)

```

### <center> **1.5.2. Calcula la complejidad del algoritmo** </center>

Para calcular la complejidad del algoritmo tenemos que tener en cuenta los siguientes aspectos.

**Complejidad de Espacio:**

num_actors: O(num_actors)
num_tomas: O(num_tomas)
num_days: O(num_days)
population_size: O(population_size * num_days)
generations: O(population_size * num_days)


**Complejidad de Tiempo:**

calculate_cost(tomas_per_day): O(num_tomas)   ### Como podemos observar, calculate_cost() no tiene complejidad de espacio. Su complejidad de espacio es la que necesite en cada iteración. (Lo mismo se aplica para el resto)

evaluate(individual): O(num_days * num_tomas)
find_optimal_schedule: O(population_size * num_days * max_tomas_per_day)
Operaciones genéticas (mate, mutate, select): O(population_size * num_days)

Cálculo final de complejidad de tiempo: O(population_size * num_days * max_tomas_per_day) + O(population_size * num_days * max_tomas_per_day) + O(generations * population_size * num_days * max_tomas_per_day) + O(generations * population_size * num_days * max_tomas_per_day) + O(generations * num_days * max_tomas_per_day)


La complejidad final del algoritmo genético en tiempoo es (generations * population_size * num_days * max_tomas_per_day), donde generations es el número de generaciones, pupulation_size es el tamaño de la población, num_days es el número de días en el que agrupamos las tomas y max_tomas_per_day es el límite máximo de tomas por día.

### <center> **4.4.2. Calcula la complejidad del algoritmo por fuerza bruta** </center>

El algoritmo genera expresiones matemáticas para evaluar un número objetivo usando combinaciones de operadores y números. La complejidad en tiempo es alta debido a los bucles anidados que recorren todas las combinaciones posibles de operadores y números, lo que resulta en O(m! * n^2). La complejidad en espacio también es significativa debido a las grandes listas de permutaciones de números, siendo O(15,120), y las combinaciones de operadores, siendo O(1). En resumen, el algoritmo es ineficiente en términos de tiempo y espacio.

Si entramos un poco más en detalle en la complejidad de tiempo podemos concluir que:

Generación de combinaciones de operadores (signscomb):
Complejidad: O(n^2), donde "n" es el número de operadores.

Generación de permutaciones de números (nums_comb):
Complejidad: O(m!), donde "m" es el número de elementos en nums_comb.

Función find_expres:
Complejidad: O(m! * n^2), donde "m" es el número de permutaciones en nums_comb y "n" es el número de operadores.

Bucle final que invoca find_expres repetidamente para valores entre -80 y 81 teniendo en cuenta un decimal:
Complejidad: O((m! * n^2) * 160 * 10) con la prueba que hemos hecho nosotros.

Como ya hemos usado algoritmos evolutivos en el ejercicio 1 (1.3.1) para este apartado vamos a utilizar un algoritmo aleatorio. ¿Tardará más en encontrar las soluciones sin un orden?¿O se tarda menos por fuerza bruta?

```python
import itertools
import numpy as np
import random

signs = ['+', '-', '*', '/']

signscomb = list(itertools.permutations(signs, 4))

nums_comb = list(itertools.permutations(range(1, 10), 5))
print(nums_comb)
print(nums_comb)
def generate_expression(nums, signs):
    # Generate a random expression using numbers and signs without repetition
    nums_sample = random.choice(nums)
    signs_sample = random.choice(signs)
    expression = [
        str(nums_sample[0]) + signs_sample[0] + str(nums_sample[1]) + signs_sample[1] +
        str(nums_sample[2]) + signs_sample[2] + str(nums_sample[3]) + signs_sample[3] +
        str(nums_sample[4])
    ]
    return expression[0]

def find_expres(p, it_count=0):
    for _ in range(1000):  # Run 1000 iterations to find the expression
        expression = generate_expression(nums_comb, signscomb)
        if eval(expression) == p:
            return print("Numero de iteraciones:", it_count + 1, "Número:", p, "Expresión:", expression)
        it_count += 1

    return print("No existe expresión, tras 1000 iteraciones:", p, "Numero de iteraciones:", it_count)

find_expres(2)
find_expres(7)
find_expres(234)
find_expres(76)
```

Como podemos observar, el número de iteraciones baja muy notablemente. Una posible mejora de este algoritmo es ver que número o que rango de números no encuentran soluciones tras x iteraciones random y que rango si encuentra rápidamente. Consiguiendo combinar una lógica inductiva en la resolución, a partir de restricciones previas que mejoren el rendimiento.
